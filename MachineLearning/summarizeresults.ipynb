```python
X = dataset.drop("Fake", axis=1)
y = dataset["Fake"]
```

```python
def moe(C, std, n):
    er = (std/n**.5)*stats.norm.ppf(C+(1-C)/2,0,1)
    return er
```

```python

for column in X:
    if X[column].dtypes == "float64" or  X[column].dtypes == "int64":
        mean = np.mean(X[column])
        #sem = np.std(X[column], ddof=1) / np.sqrt(np.size(X[column]))
        std = X[column].std()
        n = len(X[column])
        er=moe(0.9, std, n)
        print(column, mean, er)
```
